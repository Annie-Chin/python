```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd 
```

```python
x=np.linspace(0,10,50)
sinus=np.sin(x)
cosinus=np.cos(x)
plt.plot(x, sinus, 'r-o', x, cosinus, 'g--')
plt.show()
```

```python
x=np.linspace(0,10,50)
sinus=np.sin(x)
cosinus=np.cos(x)
plt.plot(x, sinus, 'r-o', x, cosinus, 'g--')
plt.grid(True)
plt.show
```
